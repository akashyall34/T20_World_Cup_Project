# T20 World Cup Data Analysis End-to-End

```python
import pandas as pd
import json
```

```python
with open('t20_json/t20_wc_match_results.json') as f:
    data = json.load(f)
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
#treat scorecard as primary key 
df_match.rename({'scorecard': 'match_id'}, axis = 1, inplace = True)
df_match.head()
```

# Create a dictionary to store both forms of match formats to ensure a match_id would become the "primary key" to identify each match

```python
match_ids_dict = {}
for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row["match_id"]
    match_ids_dict[key2] = row["match_id"]

match_ids_dict
```

```python
df_match.to_csv('dim_match_summary.csv', index = False)
```

# Batting summary

```python
with open('t20_json/t20_wc_batting_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary'])
df_batting = pd.DataFrame(all_records)
df_batting
```

```python
#make dismall out or not_out
df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x: "out" if len(x) > 0 else "not_out")
df_batting.head()
```

```python
df_batting["match_id"] = df_batting["match"].map(match_ids_dict)
df_batting.head()
```

```python
df_batting.drop(columns=["dismissal"], inplace=True)
df_batting.head(10)
```

```python
#data cleaning
df_batting['batsmanName'] = df_batting.batsmanName.apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting.batsmanName.apply(lambda x: x.replace('xa0', ''))
df_batting.head(10)
```

```python
df_batting.to_csv('batting_summary.csv', index=False)
```

# Bowling Summary

```python
with open('t20_json/t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    allRecords = []
    for rec in data:
        allRecords.extend(rec['bowlingSummary'])
```

```python
df_bowling = pd.DataFrame(allRecords)
df_bowling.head()
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_ids_dict)
df_bowling.head()
```

```python
df_bowling.to_csv('fact_bowling_summary.csv', index=False)
```

# Process Players Info.

```python
with open('t20_json/t20_wc_player_info.json') as f:
    data = json.load(f)
```

```python
df_players = pd.DataFrame(data)
df_players.head()
```

```python
df_players['name'] = df_players.name.apply(lambda x: x.replace('â€', ''))
df_players['name'] = df_players.name.apply(lambda x: x.replace('†', ''))
df_players['name'] = df_players.name.apply(lambda x: x.replace('\xa0', ''))
df_players.head()
```

```python
df_players.to_csv('dim_players_no_images.csv', index=False)
```
